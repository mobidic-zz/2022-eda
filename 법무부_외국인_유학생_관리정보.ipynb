```python
# 파일 정비
csv_files = [
    '법무부_유학생관리정보 데이터_20220630.csv',
    '법무부_유학생 개방 데이터(21년 하반기).csv',
    '법무부_유학생 개방 데이터(21년상반기).csv',
    '법무부_유학생 개방데이터(2020년하반기).csv',
    '법무부_외국인유학생 관리정보(2020년상반기).csv',
    '법무부_국내유학중인 외국인유학생2019년하반기.csv',
    '법무부_2019년 상반기 국내 유학중인 외국인유학생(전체).csv',
    '법무부_유학생 개방 데이터(2018년 하반기)전체.csv',
    '법무부_2018년 상반기 국내 유학중인 외국인유학생 관련자료.csv',
    '유학생_개방데이터(2017년12월31일 기준.csv',
    '유학생_개방데이터_201706.csv',
    '국내유학중인 외국인유학생 데이터(2016년 12월기준).csv',
    '국내유학중인 외국인유학생 데이터(2016년 6월기준).csv',
    '정보화센터_유학생_201512.csv',
    '정보화센터_유학생_201507_최종.csv',
    '법무부_외국인유학생관리정보(2014년하반기).csv',
    '법무부_외국인유학생관리정보(2014년상반기).csv']

import os
import glob
import pandas as pd

def reload_data(csv_files, show=False, low_memory=True):
    """
    CSV 를 다시 읽어온다.
    return {(파일명, dataframe)}
    """
    df_dict = {}
    for csv in csv_files:
        csv_path = os.path.join('data/', csv)
        path, fname = os.path.split(csv_path)
        if show: display(fname)
        df = pd.read_csv(csv_path, encoding='cp949', low_memory=low_memory)
        if show: display(df.head(1))
        df_dict[fname] = df
    return df_dict

df_dict = reload_data(csv_files, show=True)
```

### 데이터에 조사시기 추가

- 조서 시기는 파일명에 포함되어 있으나 규칙 없는 데이터로 손수 추가해야함

```python
# 파일명 별 조사 시기 정보 구축, 연도, 상반기, 하반기 정보 추가

csv_files2 = [
    ('법무부_유학생관리정보 데이터_20220630.csv', '2022', '6'),
    ('법무부_유학생 개방 데이터(21년 하반기).csv', '2021', '12') ,
    ('법무부_유학생 개방 데이터(21년상반기).csv', '2021', '6'),
    ('법무부_유학생 개방데이터(2020년하반기).csv', '2020', '12'),
    ('법무부_외국인유학생 관리정보(2020년상반기).csv', '2020', '6'),
    ('법무부_국내유학중인 외국인유학생2019년하반기.csv', '2019', '12'),
    ('법무부_2019년 상반기 국내 유학중인 외국인유학생(전체).csv', '2019', '6'),
    ('법무부_유학생 개방 데이터(2018년 하반기)전체.csv', '2018', '12'),
    ('법무부_2018년 상반기 국내 유학중인 외국인유학생 관련자료.csv', '2018', '6'),
    ('유학생_개방데이터(2017년12월31일 기준.csv', '2017', '12'),
    ('유학생_개방데이터_201706.csv', '2017', '6'),
    ('국내유학중인 외국인유학생 데이터(2016년 12월기준).csv', '2016', '12'),
    ('국내유학중인 외국인유학생 데이터(2016년 6월기준).csv', '2016', '6'),
    ('정보화센터_유학생_201512.csv', '2015', '12'),
    ('정보화센터_유학생_201507_최종.csv', '2015', '6'),
    ('법무부_외국인유학생관리정보(2014년하반기).csv', '2014', '12'),
    ('법무부_외국인유학생관리정보(2014년상반기).csv', '2014', '6')]

def reload_data2(csv_files, show=False, low_memory=True):
    """
    CSV 를 다시 읽어온다.
    return {(파일명, dataframe)}
    """
    df_dict = {}
    for csv in csv_files:
        csv_path = os.path.join('data/', csv[0])
        path, fname = os.path.split(csv_path)
        if show: display(fname)
        df = pd.read_csv(csv_path, encoding='cp949', low_memory=low_memory)
        
        df['조사년월'] = ['{}{:02d}'.format(csv[1], int(csv[2])) for i in range(len(df))]
        # df['조사년월'] = pd.to_datetime(df['조사년월'], format='%Y%m')
        df_dict[fname] = df
        if show: display(df.head(1))
    return df_dict

df_dict = reload_data2(csv_files2, show=True, low_memory=False)
```

```python
# 이제 붙여보자
df = pd.concat(df_dict.values(), ignore_index=True)
df.reset_index()
df.info()
display(df.head(10))
```

```python
# 필요 없는 컬럼을 제거
del df['순번']
del df['Unnamed: 7']
df.reset_index()
df.info()
df.head(10)
```

```python
# 연령대 정보를 활용하면 2022, 2011를 제외하고 생년 정보 추가가 가능
print(df[df['생년'].isnull()]['조사년월'].unique())
print(df[df['연령대'].isnull()]['조사년월'].unique())
print(df[df['생년'].isnull() & df['연령대'].notnull()]['조사년월'].unique())
print(df[df['생년'].notnull() & df['연령대'].isnull()]['조사년월'].unique())

# 연령대, 생년이 모두 없는 정보
print('---'*10)
print(df[df['연령대'].isnull() & df['생년'].isnull()]['조사년월'].unique())
```

```python
# 연령대에서 '년대생' 제거
print(df['연령대'].unique())
df['연령대'] = df['연령대'].str.replace('년대생', '')
print(df['연령대'].unique())
```

```python
# 바뀐 연령대 정보를 생년에 추가
df.loc[df['연령대'].notnull(), '생년'] = df[df['연령대'].notnull()]['연령대']
df['생년'] = df['생년'].astype('float')
df[df['연령대'].notnull()][['생년', '연령대']].head(3)
```

```python
# 조사 기간과 생년 정보를 이용하여 연령대를 나이로 계산하여 변경, 컬럼명은 연령으로 변경
df = df.rename(columns={"연령대": "연령"})
df.loc[df['생년'].notnull(), "연령"] = (df[df['생년'].notnull()]['조사년월'].str.slice(0, 4).astype(float) - df[df['생년'].notnull()]['생년']).astype('int')

# 생년이 없는 레코드의 '연령'은 None 으로 지정
df.loc[df['생년'].isnull(), '연령'] = None

# 확인
df[df['연령'].notnull()].head(5)
```

```python
# 각 필드도 확인해보자.
print(df['성별'].unique())
print(df['국적명'].unique())
print(df['체류자격'].unique())
print(df['학교명'].unique())
print(df['체류지 시도'].unique())
print(df['체류지 시군구'].unique())
```
