```python
# 파일 정비
csv_files = [
    '법무부_유학생관리정보 데이터_20220630.csv',
    '법무부_유학생 개방 데이터(21년 하반기).csv',
    '법무부_유학생 개방 데이터(21년상반기).csv',
    '법무부_유학생 개방데이터(2020년하반기).csv',
    '법무부_외국인유학생 관리정보(2020년상반기).csv',
    '법무부_국내유학중인 외국인유학생2019년하반기.csv',
    '법무부_2019년 상반기 국내 유학중인 외국인유학생(전체).csv',
    '법무부_유학생 개방 데이터(2018년 하반기)전체.csv',
    '법무부_2018년 상반기 국내 유학중인 외국인유학생 관련자료.csv',
    '유학생_개방데이터(2017년12월31일 기준.csv',
    '유학생_개방데이터_201706.csv',
    '국내유학중인 외국인유학생 데이터(2016년 12월기준).csv',
    '국내유학중인 외국인유학생 데이터(2016년 6월기준).csv',
    '정보화센터_유학생_201512.csv',
    '정보화센터_유학생_201507_최종.csv',
    '법무부_외국인유학생관리정보(2014년하반기).csv',
    '법무부_외국인유학생관리정보(2014년상반기).csv']

import os
import glob
import pandas as pd

def reload_data(csv_files, show=False, low_memory=True):
    """
    CSV 를 다시 읽어온다.
    return {(파일명, dataframe)}
    """
    df_dict = {}
    for csv in csv_files:
        csv_path = os.path.join('data/', csv)
        path, fname = os.path.split(csv_path)
        if show: display(fname)
        df = pd.read_csv(csv_path, encoding='cp949', low_memory=low_memory)
        if show: display(df.head(1))
        df_dict[fname] = df
    return df_dict

df_dict = reload_data(csv_files, show=True)
```

### 데이터에 조사시기 추가

```python
# 파일명 별 조사 시기 정보 구축, 연도, 상반기, 하반기 정보 추가

csv_files2 = [
    ('법무부_유학생관리정보 데이터_20220630.csv', '2022', '6'),
    ('법무부_유학생 개방 데이터(21년 하반기).csv', '2021', '12') ,
    ('법무부_유학생 개방 데이터(21년상반기).csv', '2021', '6'),
    ('법무부_유학생 개방데이터(2020년하반기).csv', '2020', '12'),
    ('법무부_외국인유학생 관리정보(2020년상반기).csv', '2020', '6'),
    ('법무부_국내유학중인 외국인유학생2019년하반기.csv', '2019', '12'),
    ('법무부_2019년 상반기 국내 유학중인 외국인유학생(전체).csv', '2019', '6'),
    ('법무부_유학생 개방 데이터(2018년 하반기)전체.csv', '2018', '12'),
    ('법무부_2018년 상반기 국내 유학중인 외국인유학생 관련자료.csv', '2018', '6'),
    ('유학생_개방데이터(2017년12월31일 기준.csv', '2017', '12'),
    ('유학생_개방데이터_201706.csv', '2017', '6'),
    ('국내유학중인 외국인유학생 데이터(2016년 12월기준).csv', '2016', '12'),
    ('국내유학중인 외국인유학생 데이터(2016년 6월기준).csv', '2016', '6'),
    ('정보화센터_유학생_201512.csv', '2015', '12'),
    ('정보화센터_유학생_201507_최종.csv', '2015', '6'),
    ('법무부_외국인유학생관리정보(2014년하반기).csv', '2014', '12'),
    ('법무부_외국인유학생관리정보(2014년상반기).csv', '2014', '6')]

def reload_data2(csv_files, show=False, low_memory=True):
    """
    CSV 를 다시 읽어온다.
    return {(파일명, dataframe)}
    """
    df_dict = {}
    for csv in csv_files:
        csv_path = os.path.join('data/', csv[0])
        path, fname = os.path.split(csv_path)
        if show: display(fname)
        df = pd.read_csv(csv_path, encoding='cp949', low_memory=low_memory)
        
        df['조사년월'] = ['{}{:02d}'.format(csv[1], int(csv[2])) for i in range(len(df))]
        # df['조사년월'] = pd.to_datetime(df['조사년월'], format='%Y%m')
        df_dict[fname] = df
        if show: display(df.head(1))
    return df_dict

df_dict = reload_data2(csv_files2, show=True, low_memory=False)
```

```python
# 이제 붙여보자
df = pd.concat(df_dict.values(), ignore_index=True)
df.reset_index(drop=True, inplace=True)
df.info()
display(df.head(10))
```

```python
# 필요 없는 컬럼을 제거
del df['순번']
del df['Unnamed: 7']
df.reset_index(drop=True, inplace=True)
df.info()
df.head(10)
```

```python
# 연령대 정보를 활용하면 2022, 2011를 제외하고 생년 정보 추가가 가능
# (Moon)연령대 정보를 활용하면 2022, 2021를 제외하고 생년 정보 추가가 가능
print(df[df['생년'].isnull()]['조사년월'].unique())
print(df[df['연령대'].isnull()]['조사년월'].unique())
print(df[df['생년'].isnull() & df['연령대'].notnull()]['조사년월'].unique())
print(df[df['생년'].notnull() & df['연령대'].isnull()]['조사년월'].unique())

# 연령대, 생년이 모두 없는 정보
print('---'*10)
print(df[df['연령대'].isnull() & df['생년'].isnull()]['조사년월'].unique())
```

```python
# 연령대에서 '년대생' 제거
print(df['연령대'].unique())
df['연령대'] = df['연령대'].str.replace('년대생', '')
print(df['연령대'].unique())
```

```python
# 바뀐 연령대 정보를 생년에 추가
df.loc[df['연령대'].notnull(), '생년'] = df[df['연령대'].notnull()]['연령대']
df['생년'] = df['생년'].astype('float')
df[df['연령대'].notnull()][['생년', '연령대']].head(3)
```

```python
# 조사 기간과 생년 정보를 이용하여 연령대를 나이로 계산하여 변경, 컬럼명은 연령으로 변경
df = df.rename(columns={"연령대": "연령"})
df.loc[df['생년'].notnull(), "연령"] = (df[df['생년'].notnull()]['조사년월'].str.slice(0, 4).astype(float) - df[df['생년'].notnull()]['생년']).astype('int')

# 생년이 없는 레코드의 '연령'은 None 으로 지정
df.loc[df['생년'].isnull(), '연령'] = None

# 확인
df[df['연령'].notnull()].head(5)
```

## 범주형 데이터 클렌징

```python
# 체류자격
print(df[df['체류자격'] == '학술연구기관 특정연구자']['학교명'].unique())
print(df[df['체류자격'] == '연구유학']['학교명'].unique())
print(df[df['체류자격'] == '시간제 아르바이트']['학교명'].unique())
print(df[df['체류자격'] == '외국어연수생']['조사년월'].unique())
print(df[df['체류자격'] == '(구)교환학생']['조사년월'].unique())
print(df[df['체류자격'] == '교환학생']['조사년월'].unique())

# 가급적 최신 형태로 맞춤
# 박사유학 -> 박사과정
# 석사유학 -> 석사과정 
# 전문학사 -> 전문학사과정 
# 학사유학 -> 학사과정 
# '대학부설 어학원 연수' -> '대학부설어학원연수'
# 외국어연수 -> 외국어연수생 
# (구)교환학생, 교환학생은 최근까지 혼용하는 것 같으니 유지

df['체류자격'] = df['체류자격'].str.replace('박사유학', '박사과정')
df['체류자격'] = df['체류자격'].str.replace('석사유학', '석사과정')
df['체류자격'] = df['체류자격'].str.replace('전문학사', '전문학사과정')  
df['체류자격'] = df['체류자격'].str.replace('학사유학', '학사과정')
df['체류자격'] = df['체류자격'].str.replace('대학부설 어학원 연수', '대학부설어학원연수')
df['체류자격'] = df['체류자격'].str.replace('외국어연수', '외국어연수생')  
# df['체류자격'] = df['체류자격'].str.replace('전문학사과정과정', '전문학사과정')
print(sorted(df['체류자격'].unique()))
```

```python
# 학교명 클렌징
print(df['학교명'].unique())

# 학교명 미상 -> 미상
# NaN -> 미상

df['학교명'] = df['학교명'].str.replace('학교명 미상', '미상') 
df['학교명'] = df['학교명'].fillna('미상')

# '학교명' 데이터에 기업명 확인 - '(주)원푸드림' '(주)성주디앤디' 등 아마도 연구소 일 듯, '미상', '학교명 미상'
print(df['학교명'].unique())
```

```python
# 체류지 시도 클렌징
# '체류지 시도' 데이터에 결측치 및 이상치 확인 - '미상' '체류지미상', 'Ltd)', ' Inc)'

# Ltd) , ' Inc)' 2016년도 6월 데이터가 잘 못 입력되어있음이 확인되어 수정
display(df[df['체류지 시도'] == 'Ltd)']) # (주)브릭앤(Brick & Co.,Ltd),경기도,수원시 영통구
display(df[df['체류지 시도'] == ' Inc)']) # 주식회사 래티튜드 (Latitude, Inc),서울특별시,마포구

df.loc[df['체류지 시도'] == 'Ltd)', '학교명'] = '(주)브릭앤(Brick & Co.,Ltd)'
df.loc[df['체류지 시도'] == 'Ltd)', '체류지 시군구'] = '수원시 영통구'
df.loc[df['체류지 시도'] == 'Ltd)', '체류지 시도'] = '경기도'

df.loc[df['체류지 시도'] ==' Inc)', '학교명'] = '주식회사 래티튜드 (Latitude, Inc)'
df.loc[df['체류지 시도'] ==' Inc)', '체류지 시군구'] = '마포구'
df.loc[df['체류지 시도'] ==' Inc)', '체류지 시도'] = '서울특별시'

# df.loc[1756098, '학교명'] = '(주)브릭앤(Brick & Co.,Ltd)'
# df.loc[1818250, '학교명'] = '주식회사 래티튜드 (Latitude, Inc)' 
display(df.iloc[1756098])
display(df.iloc[1818250])

# '미상' '체류지미상' '시도 미상'
# 체류지미상 -> 미상
# 시도 미상 -> 미상
df['체류지 시도'] = df['체류지 시도'].str.replace('체류지미상', '미상')
df['체류지 시도'] = df['체류지 시도'].str.replace('시도 미상', '미상')
df['체류지 시도'] = df['체류지 시도'].fillna('미상')

# 확인
print(df['체류지 시도'].unique())
print(df['학교명'].unique())
```

```python
# 체류지 시군구 클렌징
# '체류지 시군구' 데이터에 결측치 및 공백 문자 ' '가 확인 - '시군구 미상'

print(df['체류지 시군구'].unique())
display(df[df['체류지 시군구'] == ' ']) # 세종 특별 자치시
display(df[df['체류지 시군구'] == '미상']) # 시군구 미상 -> 체류지 시도가 미상인 것들은 미상으로 작성

# 확인
display(df[df['체류지 시군구'] == ' ']['체류지 시도'].unique()) # ['세종특별자치시', '충청남도'] 인 경우에 ' ' 으로 시군구가 지정되어 있음
display(df[df['체류지 시군구'].isnull()]['체류지 시도'].unique()) # '미상', '세종특별자치시'

# 체류지 시도가 미상인 경우 '미상'으로 1차 전환
display(df.loc[df['체류지 시도'] == '미상'])
df.loc[df['체류지 시도'] == '미상']['체류지 시군구'].fillna('미상')

# 체류지 시군구가 '체류지미상' -> '미상'
df['체류지 시군구'] = df['체류지 시군구'].str.replace('체류지미상', '미상')
df['체류지 시군구'] = df['체류지 시군구'].str.replace('시군구 미상', '미상') 

# df['체류지 시군구'] = df['체류지 시군구'].str.replace(' ', '미상')
# df['체류지 시군구'] = df['체류지 시군구'].fillna('미상')

# 체류지 시도와 체류지 시군구가 같은 데이터는 
# 먼저 체류지 시군구를 ' '로 바꿔두고 체류지 시도가 '미상'인 것들만 '미상'으로 변경해두자.
display(df[df['체류지 시도'] == df['체류지 시군구']])
df.loc[df['체류지 시도']==df['체류지 시군구'], '체류지 시군구'] = ' '
df.loc[df['체류지 시도']=='미상', '체류지 시군구'] = '미상'

# 체류지 시군구가 NaN 인경우는 ' '로 변경 
display(df[df['체류지 시군구'].isnull()])
display(df[df['체류지 시군구'].isnull()]['체류지 시도'].unique()) # 세종특별자치시
display(df[df['체류지 시군구'] == ' ']['체류지 시도'].unique()) # 충청남도, 경기도, 서울특별시
df['체류지 시군구'] = df['체류지 시군구'].fillna(' ')

                               
# 확인
print(df['체류지 시군구'].unique())
```

```python
# 연령, 생년의 결측치는 0으로 지정하고 데이터 타입은 int 로 변경
# (Moon) 결측치는 평균으로 대체하는것이 좋을 것 같습니다
df['연령'] = df['연령'].fillna(0).astype('int')
df['생년'] = df['생년'].fillna(0).astype('int')

# 확인
print(df['연령'].unique())
print(df['생년'].unique())
```

```python
# 마지막으로 중복 데이터 제거해야하는데 개별 통계라서 드랍하면 안된다. ^^
# df = df.drop_duplicates() 하지마!! 
df.reset_index(drop=True, inplace=True)
```

```python
# 연도와 연령의 0 값은 NaN으로 수정 후 저장 - 시각화 및 통계는 해당 값을 사용하는게 좋을 듯
import numpy as np
df.loc[df['연령']==0, '연령'] = df[df['연령']==0]['연령'].replace(0, np.NaN).astype(float)
df.loc[df['생년']==0, '생년'] = df[df['생년']==0]['생년'].replace(0, np.NaN).astype(float)

display(df.head(10))

df.to_csv('./data/클렌징_법무부_외국인_유학생_관리정보_fin.csv', index=False)
```
